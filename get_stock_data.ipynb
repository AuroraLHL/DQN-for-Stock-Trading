```python
import yfinance as yf
yf.pdr_override()
import os
```

```python
# get stock data
def get_stock_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN','600967.SS'] 
# tickers = ['600967.SS']
start_date = "2016-12-14"
end_date = "2018-06-08"
eval_start_date = "2018-06-11"
eval_end_date = "2018-08-30"
directory = 'StockData'
if not os.path.exists(directory): # create directory if it doesn't exist
    os.makedirs(directory)

for ticker in tickers:
    data = get_stock_data(ticker, start_date, end_date)
    data.to_csv(f'{directory}/{ticker}.csv') # save data to csv file
    eval_data = get_stock_data(ticker, eval_start_date, eval_end_date)
    eval_data.to_csv(f'{directory}/{ticker}_eval.csv') # save evaluation data to csv file
    
    
```
